```python
colors = ['blue', 
          'orange', 
          'green', 
          'red', 
          'brown', 
          'pink',]
```

```python
bp = plt.boxplot(pwd1_array, 
                 patch_artist=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
    bp['caps'][2*i + 0].set(color=colors[i])
plt.xlabel('Sub County') 
plt.ylabel('No. of PWD')
plt.xticks([1, 2, 3, 4, 5, 6], ['Visual', 'Hearing', 'Mobility', 'Cognition', 'Self Care', 'Communication'])
plt.xticks(rotation=40, horizontalalignment='right')
plt.title("Box Plot - Persons with Disability in\n"+"Nairobi County in 2019")
plt.savefig("Images/Boxplot.png", bbox_inches="tight")
plt.show()
```

```python
#Violin Plot
vp = plt.violinplot(pwd1_array,
                    showmedians=True,
                    vert=False)

plt.xlabel('Sub County')
plt.yticks([1, 2, 3, 4, 5, 6], ['Visual', 
                                'Hearing', 
                                'Mobility', 
                                'Cognition', 
                                'Self_Care', 
                                'Communication'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i]) 
    
plt.legend(handles = [vp['bodies'][0], 
                      vp['bodies'][1],
                      vp['bodies'][2], 
                      vp['bodies'][3], 
                      vp['bodies'][4],
                      vp['bodies'][5]],
               
           labels = ['Visual', 
                     'Hearing', 
                     'Mobility', 
                     'Cognition', 
                     'Self_Care', 
                     'Communication'], 
           loc = 'upper left',
           bbox_to_anchor = (1,1))
```

```python
# Modify the table to only have the totals for the 6 categories of the PWD 
pwd2 = pwd[['Sub_County', 'Visual_Total', 'Hearing_Total', 'Mobility_Total', 'Cognition_Total', 'Self_Care_Total', 'Communication_Total']]
pwd2
```

```python
#Append a column to sum up no. of pwd in the various sub counties in Nairobi
pwd2['PWD_Totals'] = pwd2.sum(axis=1)
pwd2
```

```python
# Plot a Pie Chart
sub_counties = pwd2["Sub_County"]
pwd_totals = pwd2["PWD_Totals"]
colors = ["#26727B", "#65338D", "#4770B3", "#D21F75", "#3B3689", "#50AED3", "#48B24F", "#E57438", "#569DD2", "#569D79", "#58595B"]
explode = (0.1, 0.2, 0, 0, 0, 0, 0, 0, 0, 0, 0)  
plt.pie(pwd_totals, labels=sub_counties, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Persons with Disability in\n"+"Nairobi County in 2019")

plt.savefig('Images/Pie Chat.png')
plt.show()
```
